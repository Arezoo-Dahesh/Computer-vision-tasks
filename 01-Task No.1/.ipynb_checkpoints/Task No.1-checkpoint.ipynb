```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob, random
import os
```

```python
print(cv2.__version__)
```

## Pick a random image

```python
directory = r"D:\pic"
file_path_type = [directory+ r"\*.jpg", directory+r"\*.jpeg"]
images = glob.glob(random.choice(file_path_type))
random_image = random.choice(images)
```

```python
image = cv2.imread(random_image)
cv2.imshow("Image", image)
cv2.waitKey()

# avoid "NOT responding" 
cv2.destroyAllWindows()
```

```python
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_RGB)
```

## Convert image to gray-scale mode

```python
gray_scale_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Image (gray_scale)", gray_scale_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(gray_scale_img, cmap="gray")
```

## Change the color of its middle pixels to red

```python
# h = image.shape[0]
# w = image.shape[1]

# or
(h, w) = image.shape[:2]
print("Image size: ", h, w)

(c_X, c_Y) = (w // 2, h // 2)
print("The middle of the image: ", c_X, c_Y)
```

```python
# 1
image_2 = image.copy()
image_2[c_Y:c_Y+1, :] = (0, 0, 255)
cv2.imshow("Colored", image_2)
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
# 2
image_3 = image.copy()
image_3[c_Y-10:c_Y+11, :] = (0, 0, 255)
cv2.imshow("Colored", image_3)
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
# 3
image_3 = image.copy()
image_3[c_Y-50:c_Y+51, c_X-50:c_X+51] = (0, 0, 255)
cv2.imshow("Colored", image_3)
cv2.waitKey()

cv2.destroyAllWindows()
```

## The image must be flipped (On Horizontal edge)

```python
# Flip code 0 -> flip vertically
flip_img_v = cv2.flip(image_3, 0)
cv2.imshow("filp_image", flip_img_v)
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
# 90 degrees rotated
rotate_cw_90 = cv2.rotate(flip_img_v, cv2.ROTATE_90_CLOCKWISE)
cv2.imshow("Rotate image", rotate_cw_90)
cv2.waitKey()

cv2.destroyAllWindows()
```

## Save the image

```python
os.chdir(directory)
print("Before saving image:")  
print(os.listdir(directory)) 
```

```python
filename = 'final_image.jpg'
cv2.imwrite(filename, rotate_cw_90)

print(os.listdir(directory))
print('Successfully saved')
```

## change_img() function

```python
def change_img(directory, filename):
    # pick a random image
    file_path_type = [directory+ r"\*.jpg", directory+r"\*.jpeg"]
    images = glob.glob(random.choice(file_path_type))
    random_image = random.choice(images)
    image = cv2.imread(random_image)
    
    # convert image to gray-scale mode
    gray_scale_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # change the color of its middle pixels to red
    (h, w) = image.shape[:2]
    (c_X, c_Y) = (w // 2, h // 2)
    image[c_Y-50:c_Y+51, c_X-50:c_X+51] = (0, 0, 255)

    # flipped (On Horizontal edge) -> flip vertically
    flip_img_v = cv2.flip(image, 0)
    
    # 90 degrees rotated
    rotate_cw_90 = cv2.rotate(flip_img_v, cv2.ROTATE_90_CLOCKWISE)
    
    cv2.imshow("Altered image", rotate_cw_90)
    cv2.waitKey()
    cv2.destroyAllWindows()
    
    # save the image
    cv2.imwrite(filename, rotate_cw_90)
    print(os.listdir(directory))
    print('Successfully saved')
```

```python
change_img(r"D:\pic", "change_img_01.jpg")
```
